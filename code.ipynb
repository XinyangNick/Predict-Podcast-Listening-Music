# Visualization

```python
import matplotlib.pyplot as plt

train_df.hist(bins=50, figsize=(20, 15))
```

```python
# Drop the outliers

print(train_df[train_df['Episode_Length_minutes'] > 120].shape)
print(train_df[train_df['Host_Popularity_percentage'] > 100].shape)
print(train_df[train_df['Host_Popularity_percentage'] < 15].shape)
print(train_df[train_df['Guest_Popularity_percentage'] > 100].shape)
print(train_df[train_df['Number_of_Ads'] > 7].shape)

def drop_outliers(df):
    cleaned_df = df[df['Episode_Length_minutes'] <= 120]
    cleaned_df = cleaned_df[cleaned_df['Host_Popularity_percentage'] <= 100]
    cleaned_df = cleaned_df[cleaned_df['Host_Popularity_percentage'] >= 15]
    cleaned_df = cleaned_df[cleaned_df['Guest_Popularity_percentage'] <= 100]
    cleaned_df = cleaned_df[cleaned_df['Number_of_Ads'] <= 7]
    return cleaned_df

train_df_cleaned = drop_outliers(train_df_drop)
print(train_df_cleaned.shape)

train_df_cleaned.hist(bins=50, figsize=(20, 15))
```

```python
# Visualize the distribution of the Episode_Title and Podcast_Name columns
train_df_cleaned['Episode_Title'].value_counts().plot(kind='bar')
plt.title('Top 10 Episode Titles')
plt.xlabel('Episode Title')
plt.ylabel('Count')
plt.show()
train_df_cleaned['Podcast_Name'].value_counts().plot(kind='bar')
plt.title('Top 10 Podcast Names')
plt.xlabel('Podcast Name')
plt.ylabel('Count')
plt.show()
```

# Imputer Test

```python
# Get a non-null subset of the data
complete_row = train_df_cleaned.copy()
print(complete_row.columns)

# drop the id column
complete_row = complete_row.drop(columns=['id'])
print(complete_row.columns)
```

```python
# Split the data into train and test sets randomly and stratified by the target variable
from sklearn.model_selection import train_test_split

imputer_train_set, imputer_test_set = train_test_split(complete_row, test_size=0.2, random_state=42)
```

```python
# get the target variable and features
X_train = imputer_train_set.drop(columns=['Listening_Time_minutes'])
y_train = imputer_train_set['Listening_Time_minutes']
X_test = imputer_test_set.drop(columns=['Listening_Time_minutes'])
y_test = imputer_test_set['Listening_Time_minutes']
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
print(X_train.columns)
print(len(X_train.columns))

num_attributes = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_attributes = X_train.select_dtypes(include=['object']).columns.tolist()

print(num_attributes)
print(cat_attributes)
print(len(num_attributes))
print(len(cat_attributes))
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

num_pipeline = Pipeline([
    ('minmax', MinMaxScaler())
])

cat_pipeline = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attributes),
    ('cat', cat_pipeline, cat_attributes)
])
```

```python
prepared_X_train = full_pipeline.fit_transform(X_train)
print(prepared_X_train)
```
